- Función para la búsqueda Best First.

```python
# Función para calcular el costo total de una ruta.
def calculate_cost(route, graph):
    cost = 0
    for i in range(len(route) - 1):
        cost += graph[route[i]][route[i + 1]]["weight"]
    return cost

# Algoritmo de Best First Search
def best_first_search(graph, initial_node, destination_node):
    priority_queue = [(0, [initial_node])]  # Inicializa una cola de prioridad con el costo y la ruta actual
    visited = set()  # Conjunto para llevar un registro de los nodos visitados

    while priority_queue:
        cost, current_route = heapq.heappop(priority_queue)  # Extrae la ruta con el menor costo de la cola de prioridad
        current_node = current_route[-1]  # Obtiene el nodo actual

        if current_node == destination_node:  # Si llegamos al nodo de destino, devuelve la ruta
            return current_route, cost

        if current_node not in visited:
            visited.add(current_node)
            neighbors = list(graph.neighbors(current_node))

            for neighbor in neighbors:
                if neighbor not in current_route:
                    new_route = current_route + [neighbor]  # Extiende la ruta con el vecino
                    new_cost = calculate_cost(new_route, graph)  # Calcula el costo de la nueva ruta
                    heapq.heappush(priority_queue, (new_cost, new_route))  # Agrega la nueva ruta a la cola de prioridad

    return None  # Si no se encuentra una ruta, devuelve None
```

- Recorremos el grafo desde "Rochester" hasta "Chicago".

```python
initial_node = "IFalls"
destination_node = "Chicago"
best_route, best_cost = best_first_search(G, initial_node, destination_node)
```

- Imprimimos los resultados.

```python
if best_route:
    print(">> Mejor ruta:\n")
    print(best_route)
    print("\n>> Costo de la mejor ruta:\n")
    print(best_cost)
else:
    print(">> No se encontró una ruta desde {} a {}.".format(initial_node, destination_node))
```
