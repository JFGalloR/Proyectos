- Función para recorrer el grafo en profundidad.

```python
def dfs(grafo, nodo_inicial): # Declaración de nuestra función de búsqueda en profundidad.
    visitados = set()
    pila = [nodo_inicial]
    recorrido = []
    # Ciclo que se realiza mientras la pila no este vacía.
    while pila:
        # Se eliminar el primer elemento de la pila.
        nodo_actual = pila.pop(0)
        # Si el nodo no se encuentra en la lista de agregados.
        if nodo_actual not in visitados:
            # Agregamos el nodo a nuetra lista de visitados.
            visitados.add(nodo_actual)
            # Agregamos el nodo a la lista del recorrido
            recorrido.append(nodo_actual)
            # Se recorren los vecinos del nodo.
            for vecino in grafo[nodo_actual]:
                #Si no estan en la lista procedemos a agregarlos.
                if vecino not in visitados:
                    pila.append(vecino)
    # Devolvemos nuestra lista de recorrido.
    return recorrido
```

- Recorremos el grafo en anchura desde "Rochester", sin embargo podemos elegir cualquier otra ciudad.

```python
r = dfs(G, "Rochester")
```

- Imprimimos el recorrido.

```python
print(">> La busqueda en profundidad es:\n")
print(r)
```
